## Part 2: Experimental Notes

**Analysis Objective:**
Characterize the biometric diversity within this penguin species across different populations (islands) and sexes.

**Measurement Variables:**
- **Culmen Length (mm):** Length of upper jaw (bill)
- **Culmen Depth (mm):** Depth/thickness of upper jaw
- **Flipper Length (mm):** Length of flippers (wings)
- **Body Mass (g):** Weight of specimen

**Expected Patterns:**
- Sexual dimorphism (males typically larger than females)
- Island-specific morphology due to local adaptation and food availability
- Correlations between flipper length and body mass (allometry)

## Part 3: Raw Data Loading

Raw biometric measurements loaded from CSV export.

```python
# Load raw data from CSV (species-specific file)
# The CSV is pre-filtered to contain only one penguin species

species_name = "Adelie"  # Replace with Chinstrap or Gentoo for other experiments

raw_data_file = f"penguins_{species_name}.csv"
df_raw = pd.read_csv(raw_data_file)

print(f"Loaded {len(df_raw)} specimens of {species_name} penguins")
print(f"\nColumns: {list(df_raw.columns)}")
print(f"\nFirst few rows:")
print(df_raw.head())
print(f"\nData shape: {df_raw.shape}")
print(f"\nMissing values:\n{df_raw.isnull().sum()}")
```

## Part 4: Data Analysis and Processing

Perform statistical analysis, visualizations, and summary calculations.

```python
# Descriptive statistics by sex
print("=" * 70)
print("DESCRIPTIVE STATISTICS BY SEX")
print("=" * 70)

stats_by_sex = df_raw.groupby('sex')[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']].describe()
print(stats_by_sex)

# Descriptive statistics by island
print("\n" + "=" * 70)
print("DESCRIPTIVE STATISTICS BY ISLAND")
print("=" * 70)

stats_by_island = df_raw.groupby('island')[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']].describe()
print(stats_by_island)
```

```python
# Visualizations (example code; not executed)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Body mass by sex
df_raw.boxplot(column='body_mass_g', by='sex', ax=axes[0, 0])
axes[0, 0].set_title('Body Mass by Sex')
axes[0, 0].set_xlabel('Sex')
axes[0, 0].set_ylabel('Body Mass (g)')

# Plot 2: Flipper length by sex
df_raw.boxplot(column='flipper_length_mm', by='sex', ax=axes[0, 1])
axes[0, 1].set_title('Flipper Length by Sex')
axes[0, 1].set_xlabel('Sex')
axes[0, 1].set_ylabel('Flipper Length (mm)')

# Plot 3: Body mass by island
df_raw.boxplot(column='body_mass_g', by='island', ax=axes[1, 0])
axes[1, 0].set_title('Body Mass by Island')
axes[1, 0].set_xlabel('Island')
axes[1, 0].set_ylabel('Body Mass (g)')

# Plot 4: Culmen measurements by sex
df_raw.boxplot(column='culmen_depth_mm', by='sex', ax=axes[1, 1])
axes[1, 1].set_title('Culmen Depth by Sex')
axes[1, 1].set_xlabel('Sex')
axes[1, 1].set_ylabel('Culmen Depth (mm)')

plt.tight_layout()
# plt.show()
```

```python
# Calculate summary metrics for this species
analysis_results = {
    'species': species_name,
    'specimen_count': len(df_raw),
    'specimen_by_sex': df_raw['sex'].value_counts().to_dict(),
    'specimen_by_island': df_raw['island'].value_counts().to_dict(),
    
    # Overall measurements
    'mean_body_mass_g': float(df_raw['body_mass_g'].mean()),
    'std_body_mass_g': float(df_raw['body_mass_g'].std()),
    'mean_flipper_length_mm': float(df_raw['flipper_length_mm'].mean()),
    'mean_culmen_length_mm': float(df_raw['culmen_length_mm'].mean()),
    'mean_culmen_depth_mm': float(df_raw['culmen_depth_mm'].mean()),
    
    # Sexual dimorphism (if both sexes present)
    'sexual_dimorphism': {
        'male_mean_body_mass': float(df_raw[df_raw['sex'] == 'Male']['body_mass_g'].mean()),
        'female_mean_body_mass': float(df_raw[df_raw['sex'] == 'Female']['body_mass_g'].mean()),
        'dimorphism_ratio': float(df_raw[df_raw['sex'] == 'Male']['body_mass_g'].mean() / df_raw[df_raw['sex'] == 'Female']['body_mass_g'].mean()),
    }
}

print("Analysis Results:")
print(json.dumps(analysis_results, indent=2))
```
